```python
import requests
from bs4 import BeautifulSoup
from lxml import html

import csv

data_list = []
c = 0
with  open(r"C:\Users\Laptop\Desktop\Car_Prediction_System\All_links.csv","r") as link:
    Links = link.readlines()
    print("read csv links")
```

```python
for link in Links:
    print(link)
    r = requests.get(link)
```

```python
for link in Links:
    
    print(link)
    r = requests.get(link)

    
    soup = BeautifulSoup(r.content, 'lxml')

    try:
        name = soup.find(['h1']).get_text()
        print("Name =",name)
        data_list.append(name)

    except:
        data_list.append("Na")

    try:     
        price = soup.find("strong",{"class":"generic-green"}).get_text()
        print(f'PRICE = {price.strip("PKR")}')
        data_list.append(price)

    except:
        data_list.append("Na")


    tree = html.fromstring(r.content)



    try:
        model = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/table/tbody/tr/td[1]/p/a/text()')
        for i in model:
            print(f"Model = {i}")
        data_list.append(model)

    except:
        data_list.append("Na")


    try:
        mileage = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/table/tbody/tr/td[2]/p/text()')
        for i in mileage:
            print(f"Mileage = {i}")
        data_list.append(mileage)

    except:
        data_list.append("Na")

    try:
        transmission = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/table/tbody/tr/td[4]/p/a/text()')
        for i in transmission:
            print(f"Transmission = {i}")
        data_list.append(transmission)

    except:
        data_list.append("Na")

    try:
        Eng_type = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/table/tbody/tr/td[3]/p/a/text()')
        for i in Eng_type:
            print(f"Engine type = {i}")
        data_list.append(Eng_type)

    except:
        data_list.append("Na")


    try:
        register = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[2]/text()')
        for i in register:
            print(f"register in = {i}")
        data_list.append(register)

    except:
        data_list.append("Na")

    try:
        color = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[4]/text()')
        for j in color:
            print(f"color = {j}")
        data_list.append(color)

    except:
        data_list.append("Na")

    try:
        assembled = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[6]/a/text()')
        for k in assembled:
            print(f"Assembled = {k}")
        data_list.append(assembled)

    except:
        data_list.append("Na")


    try:
        ENG = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[8]/text()')
        for l in ENG:
            print(f"Engine Capacity = {l}")
        data_list.append(ENG)

    except:
        data_list.append("Na")

    try:
        bd = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[10]/a/text()')
        for m in bd:
            print(f"body type = {m}")
        data_list.append(bd)

    except:
        data_list.append("Na")

    try:
        date = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[1]/li[12]/text()')
        for m in date:
            print(f"Last Updated = {m}")
        data_list.append(date)

    except:
        data_list.append("Na")



    try:
        F1 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[1]/text()')
        for m in F1:
            print(f"Feature 1 = {m}")
        data_list.append(F1)

    except:
        data_list.append("Na")



    try:
        F2 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[2]/text()')
        for m in F2:
            print(f"Feature 2 = {m}")
        data_list.append(F2)

    except:
        data_list.append("Na")


    try:
        F3 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[3]/text()')
        for m in F3:
            print(f"Feature 3 = {m}")
        data_list.append(F3)

    except:
        data_list.append("Na")



    try:
        F4 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[4]/text()')
        for m in F4:
            print(f"Feature 4 = {m}")
        data_list.append(F4)

    except:
        data_list.append("Na")



    try:
        F5 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[5]/text()')
        for m in F5:
            print(f"Feature 5 = {m}")
        data_list.append(F5)

    except:
        data_list.append("Na")



    try:
        F6 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[6]/text()')
        for m in F6:
            print(f"Feature 6 = {m}")
        data_list.append(F6)

    except:
        data_list.append("Na")



    try:
        F7 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[7]/text()')
        for m in F7:
            print(f"Feature 7 = {m}")
        data_list.append(F7)

    except:
        data_list.append("Na")


    try:
        F8 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[8]/text()')
        for m in F8:
            print(f"Feature 8 = {m}")
        data_list.append(F8)

    except:
        data_list.append("Na")



    try:
        F9 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[9]/text()')
        for m in F9:
            print(f"Feature 9 = {m}")
        data_list.append(F9)

    except:
        data_list.append("Na")



    try:
        F10 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[10]/text()')
        for m in F10:
            print(f"Feature 10 = {m}")
        data_list.append(F10)

    except:
        data_list.append("Na")


    try:
        F11 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[11]/text()')
        for m in F11:
            print(f"Feature 11 = {m}")
        data_list.append(F11)

    except:
        data_list.append("Na")


    try:
        F12 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[12]/text()')
        for m in F12:
            print(f"Feature 12 = {m}")
        data_list.append(F12)

    except:
        data_list.append("Na")



    try:
        F13 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[13]/text()')
        for m in F13:
            print(f"Feature 13 = {m}")
        data_list.append(F13)

    except:
        data_list.append("Na")


    try:
        F14 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[14]/text()')
        for m in F14:
            print(f"Feature 14 = {m}")
        data_list.append(F14)

    except:
        data_list.append("Na")



    try:
        F15 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[15]/text()')
        for m in F15:
            print(f"Feature 15 = {m}")
        data_list.append(F15)

    except:
        data_list.append("Na")



    try:
        F16 = tree.xpath('/html/body/div[6]/div/div[1]/div[2]/div[1]/div/ul[2]/li[16]/text()')
        for m in F16:
            print(f"Feature 16 = {m}")
        data_list.append(F16)

    except:
        data_list.append("Na")



    with open(r"C:\Users\Laptop\Desktop\CAR_Prediction_System_DATASET.csv", "a", encoding='UTF8',newline="") as f:
        writer = csv.writer(f)
        writer.writerow(data_list)
        data_list.clear()
        c+=1
        print(c)
```
